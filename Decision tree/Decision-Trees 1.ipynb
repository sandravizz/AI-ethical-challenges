
# Decision tree example 1

We will use the scikit-learn library to build the model and use the iris dataset which is already present in the scikit-learn library or we can download it from this [link](https://www.askpython.com/python/examples/decision-trees). The dataset contains three classes: Iris Setosa, Iris Versicolour, Iris Virginica with the following attributes: sepal length, sepal width, petal length, petal width

Objective: We have to predict the class of the iris plant based on its attributes.

## Libraries

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.datasets import load_iris
```

## Data

Simple data checks 

```python
iris = load_iris()
```

```python
print(iris.feature_names) 
```

```python
print(iris.target_names) 
```

```python
print(iris.target)
```

```python
print(iris.data.shape)
```

### Remove the Labels
Now, we will remove the elements in the 0th, 50th, and 100th position. 0th element belongs to the Setosa species, 50th belongs Versicolor species and the 100th belongs to the Virginica species.

This will remove the labels for us to train our decision tree classifier better and check if it is able to classify the data well.

```python
removed = [0,50,100]
new_target = np.delete(iris.target,removed)
new_data = np.delete(iris.data,removed, axis=0) 
```

```python
print(new_target)
```

```python
print(new_data.shape)
```

### Train the Decision Tree Classifier

The final step is to use a decision tree classifier from scikit-learn for classification.

```python
# defining decision tree classifier
clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 4) 

# train data on new data and new target
clf.fit(new_data,new_target) 

# assign removed data as input
prediction = clf.predict(iris.data[removed]) 
```

Now, we check if our predicted labels match the original labels

```python
print("Original Labels",iris.target[removed])
print("Labels Predicted",prediction)
```

## Visualisation

To plot the decision tree:

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize = (15, 8))

tree.plot_tree(clf) 
plt.show()
```
